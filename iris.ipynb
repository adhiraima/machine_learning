Manual Modelling

```python
def petal_length_model(petal_length):
    if petal_length < 2.2:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
    

def petal_width_model(petal_width):
    if petal_width < 0.8:
        return 0
    elif petal_width < 1.6:
        return 1
    else:
        return 2
```

```python
petal_width_manual = [petal_width_model(val) for val in X_train[:, 3]]
petal_length_manual = [petal_length_model(val) for val in X_train[:, 2]]

import numpy as np

result_manual_width = np.mean([petal_width_manual == y_train])
result_manual_length  = np.mean([petal_length_manual == y_train])

print(f"Manual model for petal widht: {result_manual_width * 100:.2f}% \nManual model for petal length: {result_manual_length * 100:.2f}%")
```
